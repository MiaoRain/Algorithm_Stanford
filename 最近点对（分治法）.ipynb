```python
def closestPair(Px, Py):
    if len(Px) <= 3:
        return brute(Px)

    mid = len(Px) / 2
    # get left and right half of Px 
    q, r = Px[:mid], Px[mid:]
    
    # sorted versions of q and r by their x and y coordinates 
    Qx, Qy = [x for x in q if Py and x[0] <= Px[-1][0]], [x for x in q if x[1] <= Py[-1][1]]
    Rx, Ry = [x for x in r if Py and x[0] <= Px[-1][0]], [x for x in r if x[1] <= Py[-1][1]]
    (p1, q1) = closestPair(Qx, Qy)
    (p2, q2) = closestPair(Rx, Ry)
    d = min(dist(p1, p2), dist(p2, q2))
    mn = min((p1, q1), (p2, q2), key=lambda x: dist(x[0], x[1]))
    (p3, q3) = closest_split_pair(Px, Py, d, mn)
    return min(mn, (p3, q3), key=lambda x: dist(x[0], x[1]))
```

```python
def merge_sort(alist):
    n = len(alist)
    if n <= 1:
        return(alist)
    
    mid = n // 2
    
    # 依次将 alist 中的元素对半分，分别放在 left_li 和 right_li 中
    # 递归 merge_sort 函数，直到列表中只有一个元素时，返回该元素，并进行下面的 while 循环
    
    left_li = merge_sort(alist[:mid])
    right_li = merge_sort(alist[mid:])
    left_pointer, right_pointer = 0, 0
    result = []
    
    # while 循环是将每一步中的已排好序的两个 list l_1 和 l_2 合并到 result 列表并排序，直到其中一个 list l_1 (或 l_2)全部并到 result 列表时终止循环
    
    while left_pointer < len(left_li) and right_pointer < len(right_li):
        if left_li[left_pointer] < right_li[right_pointer]:
            result.append(left_li[left_pointer])
            left_pointer += 1
        else:
            result.append(right_li[right_pointer])
            right_pointer += 1
            
    # 把另一个 list l_2中余下的元直接添加到 result 列表的最后（注：l_2 列表中的元都已排好序）
    
    result = result + left_li[left_pointer:]
    result = result + right_li[right_pointer:]
    
    return(result)
```

One dimensional case.

思路：

递归起点：$n=1$ 时，距离规定为无穷。$n=2$ 时，距离就是坐标差的绝对值。

分而治之：
当 $n>2$ 时，将其排序后等分成两个集合，递归求解左边的距离最小值和右边的距离最小值，再去计算可能存在的被分割的中间两个点的距离，比较得出其中的最小值。

```python
def CP1(P):
    import math
    n = len(P)
    
    if n == 2:
        return( abs(P[0] - P[1]) )
    elif n == 1:
        return(math.inf)
    
    else:
        mid = n // 2
        Q = sorted(P)
        left_li = Q[:mid]
        right_li = Q[mid:]

        
        delta_left = CP1(left_li)
        delta_right = CP1(right_li)

        left_max = max(left_li)
        right_min = min(right_li)

        delta_middle = abs(right_min - left_max)
        delta = min([delta_left, delta_right, delta_middle])
        return(delta)  
```

```python
def distance(a,b):
    dis = math.sqrt((a[0]-b[0])**2+(a[1]-b[1])**2)
    return(dis)


def CP2(P):
    import math
    n = len(P)
    
    if n == 2:
        return(distance(P[0], P[1]))
    elif n == 1:
        return(math.inf)    
    else:
        Q = sorted(P, key = lambda x:x[0])
        mid = n // 2
        left_li = Q[:mid]
        right_li = Q[mid:]
        
        delta_left = CP2(left_li)
        delta_right = CP2(right_li)
        delta = min([delta_left, delta_right])
        
        S = [i for i in Q if i[0] >= Q[mid][0] - delta and i[0] <= Q[mid][0] + delta]
        sortS_y = sorted(S, key = lambda x:x[1])
        
        best = delta
        for x in range( len(sortS_y) - 1 ):
            for y in range(1, min([7, len(sortS_y) - x])):
                if distance(sortS_y[x], sortS_y[x + y]) < delta:
                    best = distance(sortS_y[x], sortS_y[x + y])
        return(best)                
```
